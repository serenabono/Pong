```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import plotly
import plotly.io as pio
pio.renderers.default = 'iframe'
import plotly.express as px
plotly.offline.init_notebook_mode(connected=True)
import seaborn as sns
import matplotlib as mpl
import re
```

```python
import pandas as pd
import json
import os
import csv 
import pickle

data = {}
name = {}
for path, directories, files in os.walk('.'):
    if "_trial_" in path:
        for file in files:
            if "}.pkl" in file:
                file_name = f"{path}/{file}"
                path = path.split("/")[1]
                agent = path.split("_")[3]
                exploration_strategy = path.split("_")[4]
                grid = path.split("_")[5]
                bar = "_".join(path.split("_")[6:8])
                noise = "".join(path.split("_")[8])
                if agent not in data:
                    data[agent] = {}
                    name[agent] = {}
                if exploration_strategy not in data[agent]:
                    data[agent][exploration_strategy] = {}
                    name[agent][exploration_strategy] = {}
                if grid not in data[agent][exploration_strategy]:
                    data[agent][exploration_strategy][grid] = {}
                    name[agent][exploration_strategy][grid] = {}
                if bar not in data[agent][exploration_strategy][grid]:
                    data[agent][exploration_strategy][grid][bar] = {}
                    name[agent][exploration_strategy][grid][bar] = {}
                if noise not in data[agent][exploration_strategy][grid][bar]:
                    data[agent][exploration_strategy][grid][bar][noise] = {}
                name[agent][exploration_strategy][grid][bar][noise] = file_name
                print(f"{file_name}\n")
                with open(file_name, 'rb') as file:
                    loaded_dict = pickle.load(file)
                data[agent][exploration_strategy][grid][bar][noise] = loaded_dict
```

```python
import numpy as np
import scipy
```

```python
# Compute Spearman correlation coefficient
def test_correlation(x,y):
    return scipy.stats.spearmanr(x, y)
```

```python
def test_significance(x,y):
    return scipy.stats.wilcoxon(x,y)
```

```python
contents_lg = {}

contents_lg['blue'] = []
contents_lg['red'] = []
contents_lg['white'] = []
contents_lg['L'] = []
contents_lg['G'] = []
gap_lgs = []
names = []

for agent in data:
    for exploration_strategy in data[agent]:
        for grid in data[agent][exploration_strategy]:
            for bar in data[agent][exploration_strategy][grid]:
                for noise in data[agent][exploration_strategy][grid][bar]:
                    for EPOCH in data[agent][exploration_strategy][grid][bar][noise]:
                        lg = data[agent][exploration_strategy][grid][bar][noise][EEPOCH]
                        all_states = np.sum(lg[0:3])
                        contents_lg['red'].append(lg[0]/all_states)
                        contents_lg['blue'].append(lg[1]/all_states)
                        contents_lg['white'].append(lg[2]/all_states)
                        contents_lg['L'].append(lg[3])
                        contents_lg['G'].append(lg[4])
                        var = (lg[4]-lg[3])/(lg[4]+lg[3])
                        gap_lgs.append(var)
                        names.append(name[agent][exploration_strategy][grid][bar][noise])
```

```python
test_significance(gap_lgs, contents_lg['blue']), test_significance(gap_lgs, contents_lg['red']), test_significance(gap_lgs, contents_lg['white'])
```

```python
test_correlation(gap_lgs, contents_lg['blue']), test_correlation(gap_lgs, contents_lg['red']), test_correlation(gap_lgs, contents_lg['white'])
```

```python
import pandas as pd
pd.DataFrame({"names":names,"G-L":gap_lgs,"L":contents_lg['L'],"G":contents_lg['G'],"blue":contents_lg['blue'],"red":contents_lg['red'],"white":contents_lg['white']}).to_csv("statistics_states_occupancy.csv", index=False)
```
